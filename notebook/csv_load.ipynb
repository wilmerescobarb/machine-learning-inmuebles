```python
import pandas as pd
```

```python
#conda install -c anaconda psycopg2

from sqlalchemy import create_engine

DATABASES = {
       'db':{
        'NAME': 'bd_ml',
        'USER': 'postgres',
        'PASSWORD': 'postgres1',
        'HOST': 'localhost',
        'PORT': 5432,
    }
}

db= DATABASES['db']


prod_engine_string = "postgresql+psycopg2://{user}:{password}@{host}:{port}/{database}".format(
    user = db['USER'],
    password = db['PASSWORD'],
    host = db['HOST'],
    port = db['PORT'],
    database = db['NAME'],
)

prod= create_engine(prod_engine_string)
```

```python
lineas=pd.read_excel("lineas-de-subte.xlsx")
lineas
```

```python
lineas.to_sql(name="subte",con=prod,schema="public",if_exists = 'replace', method="multi",chunksize = 1000,index=False)
```
